```python
import pandas as pd
import matplotlib.pyplot as plt

# Örnek veri oluşturma
data = pd.DataFrame({
    'period': ['School', 'Summer', 'School', 'Summer'],
    'IslemTutarı': [100, 200, 150, 250]
})

# Dönemlere göre toplam harcamayı gruplama
period_spending = data.groupby('period')['IslemTutarı'].sum()

# Grafiği görselleştirme
plt.figure(figsize=(8, 5))
period_spending.plot(kind='bar', color=['blue', 'orange'])
plt.title('Total Spending by Period')
plt.xlabel('Period')
plt.ylabel('Total Spending (TRY)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
